```python
LB_Missing_na_with_mean = data["LB"].fillna(data["LB"].mean(), inplace=True)
AC_Missing_na_with_mean = data["AC"].fillna(data["AC"].mean(), inplace=True)
DS_Missing_na_with_mean = data["DS"].fillna(data["DS"].mean(), inplace=True)
DP_Missing_na_with_mean = data["DP"].fillna(data["DP"].mean(), inplace=True)
MLTV_Missing_na_with_mean = data["MLTV"].fillna(data["MLTV"].mean(), inplace=True)
Width_Missing_na_with_mean = data["Width"].fillna(data["Width"].mean(), inplace=True)
Tendency_Missing_na_with_mean = data["Tendency"].fillna(data["Tendency"].mean(), inplace=True)
NSP_Missing_na_with_mean = data["NSP"].fillna(data["NSP"].mean(), inplace=True)
```

```python
Q1 = data["LB"].quantile(0.25)
Q3 = data["LB"].quantile(0.75)
print("LB quantile:",Q1, Q3)
```

```python
IQR = Q3-Q1
lower_bound = Q1-1.5*IQR
upper_bound = Q3+1.5*IQR
print("IQR:",IQR)
print("lower_bound:",lower_bound)
print("upper_bound:",upper_bound)   
```

```python
data[data["LB"]<lower_bound]
```

```python
data[data["LB"]>upper_bound]
```

```python
print(data["LB"].quantile(0.05))
print(data["LB"].quantile(0.95))
```

```python
# treatment of outliers by replacing the values
# the values lesser than 119 are replaced by 119 and 
# the values greater than 149 are replaced by 149.
data["LB"] = np.where(data["LB"]<119,119,data["LB"])
data["LB"] = np.where(data["LB"]>149,149,data["LB"])
```
